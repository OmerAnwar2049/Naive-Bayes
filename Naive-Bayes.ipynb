## Naive Bayes Training Function



Returns a dictioanary containing all the unique words present in all the classes

```python
def extractVocablury(C):
  
  D = {}
  for i in  C:                  # For loop iterating each class
    for j in i:                 # For loop iterating each text (sentence in class)
      for k in j:               # For loop iterating each word in that specific text
        if k not in D.keys():   # If the specific word is not already in a dictionary
          D[k]=1                # Append new unique word.
  return D
```

```python
def countWords(c):
  count = 0;
  for i in c:       # For loop iterating a specific class (real or fake)
    count+=len(i)   # Length of each text (sentence) is added up
  return count      # Return total length of words in this class
```

THis function combines the text of a specific class to form its specific document.
1. The document is a dictionary which contains all the words and their instances as values.

```python
def concatenateTexts(c):
  D = {}
  for i in c:
    for j in i:
      if j not in D:
        D[j] = 1      # If not in dict
      else:
        D[j]+=1       # Increment if word is already present.
  return D
```

This function counts the total number of times a word from the vocablury appears in that specific documnet of class and returns the dictionary of instance of each word.
This dict is later used to form the instance of Ni used for conditional probability.

```python
def countTokens(vocab,doc):
  D = {}
  for w in vocab.keys():  # Loop iterating all unique words of vocablury
    if w in doc.keys():   # If word in vocablury also present in loop
      D[w] = doc[w]       # Store the number of instances stored above
    else:
      #
      D[w] = 0            # This condition ensures a error will not appear when iterating vocablury loop in main function to avoid key error
  return D
```

TrainMultinomaialNB()

Implemented exactly as shown in assignmennt.

1. Recieves classes and all the texts
2. For each class prior,and conditional probs are calculated.
3. Ni calulated using the countTokens function for optimisation.
4. 


```python
def TrainMultinomialNB(C,T):
  V = extractVocablury(C)
  N = len(T)

  cond_prob = dict()
  for i in V:
    cond_prob[i] = dict()

  prior={}

  for c in range(len(C)):

    Nc = len(C[c])                # Len of the training texts in class c
    Nw = countWords(C[c])         # Words in all text of the class c
    prior[c] = float(Nc/N)        # Storing Prior
    doc = concatenateTexts(C[c])  # Containing all words of class c
    Ni = countTokens(V,doc)       # Tokens of each word of vocablury in the class (loops over vocablry words in the function)

    for w in V.keys():
      cond_prob[w][c] = (Ni[w] + 1) / (Nw + len(V))   #Storing Conditional Prob for every word.

  return (V,prior,cond_prob)
    
```

## Naive Bayes Test Function

This fucntion simply recieves the rewuired text news, extracts the words that are in vocablury and news and sends a list of these words back.

```python
def extractWords(vocab,text):
  words = []                  # List of words present in both V and the text
  for i in text:
    if i in vocab.keys():     # If word in both vocab and text
      words.append(i)         # Appending the word
  return words
```

Thuis function is implemented exactly as required in the paper.
1. The function recieves classes,V,prior,conditional prob and one test text(news) file.

2. The loop runs ovet all the classes and the firstly the log of prior is added to that score. Then the conditional probs of the test words in the classes is added.

3. The value Real or Fake is returned based on wether score 0 was greater or score 1.

```python
def applyMultinomialNB(C,V,prior,condprob,t):
  
  W = extractWords(V,t)

  score = {}

  for c in range(len(C)):

    score[c] = math.log(prior[c])             # Storing Class prior

    for word in W:
      score[c] += math.log(condprob[word][c]) # Adding Conditional prob of each word.

  
  if score[0]>score[1]:
    return "Real"       # Return Real if class 0 score is greater
  else:
    return "Fake"       # Return Fake if class 1 score is greater
```

## Measuring Criteria

We will first run all the test set and fill the confusuin matrix to fill the four fields namely True negative,False negative,False Positive,True Positve

1. First train the rewuired set and then run the Test function on all the test texts for that traing set.
2. Fill the 2x2 confusion matrix accordingly for later being used in precision and recall and to find F1 of the model.
3. This has to be done for all the 4 types of datasets and their tests sets.
4. Currently we will be using for demo in this part the simple dataset with stop words and its test cases.

Confusion Matrix:
This basically contains the True negative,False negative,False Positive,True Positve.

1. True Negative: Stored in index [0][0] of the matrix.This is icremented when the class predicted is fake and the true class was also fake.
2. True Positive: Stored in index [1][1] of the matrix.This is icremented when the class predicted is real and the true class was also real.
3. False Negative: Stored in index [1][0] of the matrix.This is icremented when the class predicted is fake and the true class was real.
4.  False Positive: Stored in index [0][1] of the matrix.This is icremented when the class predicted is real and the true class was fake.

Accuracy:
This is calculated by adding the totla number of true real and true fake news dtected by out model and dividing them bu the total number of texts in the test case.

Precision:
This is calculated by dividing true positives given by out system and dividing it by the sum of true posiitves and false positives given by our system.

Recall:
This is calculated by dividnge true positive by the sum of true and false positeves.Menaing the total number of actually positive texts.

F1:
This is the balance between recall and precision and is calculated using the formula :  2 * ((precision*recall)/(precision+recall))


def Model_Testing()

1. This function recieves the Classes,Vocablury,conditional prob and Tests. All the fields required for the applyMultinomial Test function. 
2. It then runs the Tests which contains all the tests of classes and send each text (news) to the apply multinomial function.
3. Since this is a binary classifier the value returned is either real or fake.
4. The values for true poitive ,true negative,false positive,false negative are updated in the confusion matrix accordingly from the results returnded by the apply function.
5. The values of precision,recall,accuracy and f1 are calculated after all the tests have been iterated.




```python
def Model_Testing(C,V,prior,condprob,T):

  c_matrix = np.array([[0, 0], [0, 0]])

  accurate = 0

  for c in range(len(T)):
    for t in T[c]:
      result = applyMultinomialNB(C,V,prior,condprob,t)

      # FOR REAL
      if c == 0:

        if result == "Real":            # FILLING THE TRUE POSITIVE NUMBER FROM TESTS SET
          c_matrix[1][1] += 1

        elif result == "Fake":          # FILLING THE FALSE NEGATIVE NUMBER FROM TESTS SET
          c_matrix[1][0] += 1

      # FOR FAKE
      else:

        if result == "Real":            # FILLING THE FALSE POSITIVE NUMBER FROM TESTS SET
          c_matrix[0][1] += 1

        elif result == "Fake":          # FILLING THE TRUE NEGATIVE NUMBER FROM TESTS SET
          c_matrix[0][0] += 1        


  sum =   c_matrix[0][0]+c_matrix[0][1]+c_matrix[1][0]+c_matrix[1][1]

  # Calculating Accuracy
  accuracy = ((c_matrix[0][0]+c_matrix[1][1])/sum)*100

  # Calculating Precision
  precision = c_matrix[1][1] / (c_matrix[1][1] + c_matrix[0][1])

  # Calculating Recall
  recall = c_matrix[1][1] / (c_matrix[1][1] + c_matrix[0][0])

  # Calculating F1
  f1 = 2 *  ((precision*recall)/(precision+recall))

  print("Accuracy = ",accuracy,"%")
  print("Precision = ",precision)
  print("Recall = ",recall)
  print("F1 = ",f1)
  print()
  return (accuracy,precision,recall,f1)

```
